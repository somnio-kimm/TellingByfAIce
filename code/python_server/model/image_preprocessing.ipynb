```python
import os
from PIL import Image, ExifTags, ImageOps

def correct_exif_orientation(image_path, save_path=None):
	img = Image.open(image_path)

	try:
		# EXIF Orientation 키 찾기
		for key, val in ExifTags.TAGS.items():
			if val == 'Orientation':
				orientation_key = key
				break

		exif = img._getexif()
		if exif is not None:
			orientation = exif.get(orientation_key)
			print(orientation)

			# EXIF 값에 따라 회전/반전 처리
			if orientation == 2:
				img = ImageOps.mirror(img)
			elif orientation == 3:
				img = img.rotate(180, expand=True)
			elif orientation == 4:
				img = ImageOps.flip(img)
			elif orientation == 5:
				img = ImageOps.mirror(img).rotate(-90, expand=True)
			elif orientation == 6:
				img = img.rotate(-90, expand=True)
			elif orientation == 7:
				img = ImageOps.mirror(img).rotate(90, expand=True)
			elif orientation == 8:
				img = img.rotate(90, expand=True)

	except Exception as e:
		print(f"EXIF 처리 중 오류: {e}")

	if save_path:
		img.save(save_path)
		print(f"정방향으로 저장 완료: {save_path}")
	else:
		return img
```

```python
dir_img = "../data/downloaded/test/image"
dir_list = ["anger", "happy", "panic", "sadness"]
for dir in dir_list:
	path_dir = os.path.join(dir_img, dir)
	for file in os.listdir(path_dir):
		if file.lower().endswith((".jpg", ".jpeg", ".png")):
			path_img = os.path.join(path_dir, file)
			correct_exif_orientation(image_path=path_img, save_path=path_img)
```
